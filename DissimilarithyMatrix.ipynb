```python
import numpy as np
from scipy.spatial.distance import pdist, squareform
data = np.array([
    [1, 2],
    [3,4],
    [5,6]
])
condensed_distance_matrix = pdist(data, metric='euclidean')

dissimilarity_matrix = squareform(condensed_distance_matrix)

print("Dissimilarity Matrix (Euclidean):")
print(dissimilarity_matrix)
```

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform
data = np.array([
    [1, 2],
    [3,4],
    [5,6]
])
condensed_distance_matrix = pdist(data, metric='Jaccard')

dissimilarity_matrix = squareform(condensed_distance_matrix)

print("Dissimilarity Matrix (Jaccard):")
print(dissimilarity_matrix)
```
