```python
list = input().split()
list = [int(x) for x in list]

i = 0
while i < len(list):
    j = i + 1
    while j < len(list):
        if list[i] > list[j]:
            list[i], list[j] = list[j], list[i]
        j = j + 1
    i = i + 1

print(list)
print(list[::-1])
```
